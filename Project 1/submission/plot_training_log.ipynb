```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pickle

W = []
LOG = []
# SGD
with open(r'C:\MFE\MFE Sem 3\CSE 426\CSE_426\Project 1\project1_release\data\SGD_outcome.pkl', 'rb') as f:
    (w, log) = pickle.load(f)
W.append(w)
LOG.append(log)
# batch GD
with open(r'C:\MFE\MFE Sem 3\CSE 426\CSE_426\Project 1\project1_release\data\batch_outcome.pkl', 'rb') as f:
    (w, log) = pickle.load(f)
W.append(w)
LOG.append(log)
# Newton
with open('./data/Newton_outcome.pkl', 'rb') as f:
    (w, log) = pickle.load(f)
W.append(w)
LOG.append(log)
```

```python
plt.figure()
for idx in range(len(W)):
    print(idx)
    w = W[idx]
    log = LOG[idx]
    
    x = np.array(range(1,len(log)+1))
    training_loss = [e[0] for e in log]
    # validation_loss = [e[1] for e in log]
    test_loss = [e[1] for e in log]
#     print (x)
#     print (training_loss)
    if idx == 0:
        plt.plot(x, training_loss, 'y-', label='SGD_training_loss')
        plt.plot(x, test_loss, 'b-', label='SGD_test_loss')
    # plt.plot(x, test_loss, 'r-', label='test_loss')
    if idx == 1:
        plt.plot(x, training_loss, 'r-', label='Batch_training_loss')
        plt.plot(x, test_loss, 'g-', label='Batch_test_loss')
    if idx == 2:
        plt.plot(x, training_loss, 'c-', label='Newton_training_loss')
        plt.plot(x, test_loss, 'k-', label='Newton_test_loss')
    plt.legend()

    
```
